```python
import numpy as np
import pandas as pd 
import graphlab as gl
from sklearn import preprocessing
```

```python
df = pd.read_csv("C:\Users\student04\Desktop\santander proiject\GraphLabData.csv", dtype={'indrel_1mes': str, 'conyuemp': str})
```

```python
df.dtypes
#We do not want columns to be of type object because we can not put them into SFrame later when using the algorithms
#The data types are also too large
```

```python
#We change the objects and complete preprocessing
for f in df.columns:
    if df[f].dtype == 'object':
        print(f)
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(df[f].values))
        df[f] = lbl.transform(list(df[f].values))
```

```python
for x in df.columns:
    if df[x].dtypes == 'int64':
        df[x] = df[x].astype('int32')
    if df[x].dtypes == 'float64':
        df[x] = df[x].astype('float32')
#Check out this double if loop Lawrence I know you're loving this
```

```python
df.dtypes
```

```python
# we want to have 3 columns (ncodpers, products, ratings)
# to implement the similarity algorithm of Graphlab so
# we continue like that. First we create a dataframe with the products
products = df[['ind_ahor_fin_ult1','ind_aval_fin_ult1','ind_cco_fin_ult1','ind_cder_fin_ult1','ind_cno_fin_ult1','ind_ctju_fin_ult1','ind_ctma_fin_ult1','ind_ctop_fin_ult1','ind_ctpp_fin_ult1','ind_deco_fin_ult1','ind_deme_fin_ult1','ind_dela_fin_ult1','ind_ecue_fin_ult1','ind_fond_fin_ult1','ind_hip_fin_ult1','ind_plan_fin_ult1','ind_pres_fin_ult1','ind_reca_fin_ult1','ind_tjcr_fin_ult1','ind_valo_fin_ult1','ind_viv_fin_ult1','ind_nomina_ult1','ind_nom_pens_ult1','ind_recibo_ult1']]
```

```python
products.head()
```

```python
# getting rid of the last row which had only nulls
products = products.loc[:931435]
```

```python
# changing the floiats into int
for x in products.columns:
    if products[x].dtypes == 'float32':
        products[x] = products[x].astype('int32')
```

```python
# checking the data types of our product columns
products.dtypes
```

```python
# bringing the products into the format we need to then apply the graphlab 
# similarity algorithm
stack_products = products.stack().to_frame()
```

```python
# creating a dataframe with the userids or ncodpers
users = df['ncodpers']
```

```python
users.head().to_frame()
```

```python
# creating an array with the userids
a = np.array(users)
```

```python
# since we have 24 products we continue like that
values = np.repeat(a,24)
usersDF = pd.DataFrame(values)
```

```python
usersDF
```

```python
stack_prod = stack_products.reset_index()
```

```python
stack_prod
```

```python
# now we need to concatenate the users and products like this
results1 = pd.concat([usersDF,stack_prod], axis=1)
results1
```

```python
results1.columns = ['ncodpers', 'whatever', 'product', 'Y/N']
```

```python
results1.head()
```

```python
results = results1.loc[:22354463]
```

```python
# putting our dataframe into the final graphlab format
train_set = results[['ncodpers', 'product', 'Y/N']]
```

```python
train_set
```

```python
train_set2 = gl.SFrame(train_set)
```

```python
train_set2
```

```python
# trying the pearson similarity algorithm
item_sim_model = gl.item_similarity_recommender.create(train_set2, user_id='ncodpers', item_id='product', target='Y/N', similarity_type='pearson')
```

```python
# recommendations

item_sim_recomm = item_sim_model.recommend(users=range(657810, 658000), k=7)
item_sim_recomm.print_rows(num_rows=50)
```
